```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Replace 'your_file.csv' with the path to your dataset
df = pd.read_csv('C:/Users/syeda_ojd0o2c/Documents/Project 3/daily-website-visitors.csv')
df.head()
```

```python
#Rename the column
df.rename(columns= {"Row":"ID"}, inplace=True)
df.columns
```

```python
# Display the shape of the dataset
print(df.shape)
```

```python
# Display basic information about the dataset
print(df.info())
```

```python
# Display basic statistics
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())

# Fill or drop missing values as necessary
df = df.dropna()  # or df.fillna(method='ffill')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Page.Loads'], kde=True)
plt.title('Distribution of Page Loads')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Page.Loads'], label='Page Loads')
plt.plot(df['Date'], df['Unique.Visits'], label='Unique Visits')
plt.xlabel('Date')
plt.ylabel('Counts')
plt.title('Page Loads and Unique Visits Over Time')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Day.Of.Week', y='Page.Loads', data=df)
plt.title('Page Loads by Day of the Week')
plt.show()
```
